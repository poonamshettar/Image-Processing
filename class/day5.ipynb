```python
class complexn:
    def __init__(self,real,img):
        self.real=real
        self.img=img
c1=complexn(2,3)
print(f"real: {c1.real} imaginary: {c1.img}")
```

```python
class ComplexN:
    def __init__(self,real,img):
        self.real=real
        self.img=img
    def __add__(self,other):
        real_part_sum=self.real + other.real
        img_part_sum=self.img + other.img
        return ComplexN(real_part_sum,img_part_sum)
    def __sub__(self,other):
        real_part_sum=self.real - other.real
        img_part_sum=self.img - other.img
        return ComplexN(real_part_sum,img_part_sum)
    def __str__(self):
        return f"{self.real}+{self.img}i"
c1=ComplexN(2,3)
c2=ComplexN(3,2)
r=c1+c2
print(r)
r=c1-c2
print(r)
```

```python
class polynomial:
    def __init__(self,*coeffs):
        self.coeffs=coeffs
    def __repr__(self):
        return 'Poly({!r})'.format(self.coeffs)
    def __str__(self):
        return f"{self.coeffs}"
    def __add__(self,other):
        return polynomial(*(x+y for x,y in zip(self.coeffs,other.coeffs)))
p1=polynomial(1,23,4)
p2=polynomial(1,2)
print(p1+p2)
```
