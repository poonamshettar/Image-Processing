```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('a.csv')
print(df.head())
plt.plot(df['x'],df['y'],marker='*')
```

```python
plt.bar(df['x'],df['y'])
plt.plot(df['x'],df['y'])
```

```python
plt.pie(df['x'],labels=[1,2,3,4,5],autopct='%1.1f%%')
plt.legend(loc='lower left')
plt.show()
```

```python
plt.figure(figsize=(20,20),facecolor='pink')
plt.subplot(4,4,1)
plt.plot(df['x'],df['y'])
plt.title('line_plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.subplot(4,4,2)
plt.scatter(df['x'],df['y'])
plt.title('scatter')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.subplot(4,4,3)
plt.pie(df['x'],autopct='%1.1f%%',labels=[1,2,3,4,5])
plt.legend(loc='lower left')
plt.title('pie')
plt.subplot(4,4,4)
plt.bar(df['x'],df['y'])
plt.title('bar')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.savefig('Plots_all.png')
```
