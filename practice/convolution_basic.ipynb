```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Without padding

```python
image=cv2.imread('D:\CEVI\images\\tajmahal.webp',0)
kernel=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
image_h,image_w=image.shape
kernel_h,kernel_w=kernel.shape
output_img=np.zeros((image_h,image_w))
for i in range(image_h):
    for j in range(image_w):
        if(i<=image_h-kernel_h and j<=image_w-kernel_w):
            output_img[i,j]=np.sum(image[i:i+kernel_h,j:j+kernel_w]*kernel)
plt.subplot(2,2,1)
plt.imshow(image,cmap='gray')
plt.axis('off')
plt.subplot(2,2,2)
plt.imshow(output_img,cmap='gray')
plt.axis('off')
```

```python
print(image.shape)
print(output_img.shape)
```

```python
image=cv2.imread('D:\CEVI\images\\tajmahal.webp',0)
kernel=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
image_h,image_w=image.shape
kernel_h,kernel_w=kernel.shape
output_h=image_h-kernel_h+1
output_w=image_w-kernel_w+1
output_img=np.zeros((output_h,output_w))
for i in range(output_h):
    for j in range(output_w):
            output_img[i,j]=np.sum(image[i:i+kernel_h,j:j+kernel_w]*kernel)
plt.subplot(2,2,1)
plt.imshow(image,cmap='gray')
plt.axis('off')
plt.subplot(2,2,2)
plt.imshow(output_img,cmap='gray')
plt.axis('off')
```
