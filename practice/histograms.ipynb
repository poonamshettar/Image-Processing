```python
img=  cv2.imread('D:\CEVI\images\\tajmahal.webp')
b,g,r=cv2.split(img)
plt.figure(figsize=(20,20),facecolor='pink')
list=[b,g,r]
list1=['blue','green','red']
for i in range(3):
    plt.subplot(3,3,i+1)
    plt.hist(list[i].ravel(),256,[0,256])
    plt.title(f"{list1[i]}")
    plt.xlabel('intensity')
    plt.ylabel('no_of_pixels')
plt.show()
```
